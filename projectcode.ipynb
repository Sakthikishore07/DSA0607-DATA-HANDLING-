```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = r"//content//symbols_valid_meta.csv"
df = pd.read_csv(file_path)

# Convert index to datetime format
df.index = pd.to_datetime(df.index)

# Plotting time series for Nasdaq Traded
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Nasdaq Traded'], marker='o', color='blue')
plt.title('Nasdaq Traded Over Time')
plt.xlabel('Date')
plt.ylabel('Nasdaq Traded')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Plotting time series for Round Lot Size
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Round Lot Size'], marker='o', color='green')
plt.title('Round Lot Size Over Time')
plt.xlabel('Date')
plt.ylabel('Round Lot Size')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Repeat the same process for other attributes if needed
```
